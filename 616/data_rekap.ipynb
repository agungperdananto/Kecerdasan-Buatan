```python
# mapping data {tahun: penjualan}
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}
data_map
```

```python
# sorting years
years = list(data_map.keys())
years.sort()
years
```

```python
data_graph = {
    "tahun": [...],
    "penjualan": [...]
}
```

```python
data_graph = {}
for year in years:
    tahun_list = data_graph.get("tahun", [])
    penjualan_list = data_graph.get("penjualan", [])
    tahun_list.append(year)
    data_graph["tahun"] = tahun_list
    penjualan_list.append(data_map[year])
    data_graph["penjualan"] = penjualan_list

data_graph
```

```python
plt.plot(data_graph["tahun"], data_graph["penjualan"],'.--', color="C4", label="penjualan")
plt.plot(data_graph["tahun"], [i * 100 for i in range(len(years))],'--', color="#FFBBCC", label="data dummy")

plt.xlabel("tahun")
plt.ylabel("nominal dalam jutaan Rp.")
plt.title("Data Rekapitulasi")
plt.legend(loc="lower right")
plt.show()
```
