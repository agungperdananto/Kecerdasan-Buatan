```python
data_raw = {}
for dk in data_keuntungan:
    tahun = dk["tahun"]
    penjualan = dk["penjualan"]
    data_raw[tahun] = penjualan
data_raw
```

```python
data_tahun = list(data_raw.keys())
data_tahun.sort()
```

```python
data_penjualan = [data_raw[data]for data in data_tahun]
data_penjualan

data_grafik = {
    "tahun": data_tahun,
    "penjualan": data_penjualan
    # pajak, operasional , keuntungan
}
data_grafik
```

```python
plt.plot(data_grafik["tahun"], data_grafik["penjualan"], label="penjualan")
# plt.plot(data_grafik["tahun"], data_grafik["pajak"], label="pajak")

plt.legend(loc="upper left")
plt.xlabel("tahun")
plt.ylabel("dalam jutaan Rp")
plt.title("DATA REKAPITULASI TAHUNAN")
plt.show()
```
