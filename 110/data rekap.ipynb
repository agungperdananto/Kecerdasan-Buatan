```python
# data_graph = {
#     "years": [],
#     "sales": []
# }
```

```python
# mapping data {tahun: penjualan}
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}

data_map
```

```python
# sorting years
years = list(data_map.keys())
years.sort()
years
```

```python
# finalize data
data_graph = {}
for year in years:
    year_list = data_graph.get("years", [])
    sale_list = data_graph.get("sales", [])
    year_list.append(year)
    sale_list.append(data_map[year])
    data_graph["years"] = year_list
    data_graph["sales"] = sale_list
data_graph
```

```python
plt.plot(data_graph["years"], data_graph["sales"],'.--', label="Penjualan")
plt.plot(data_graph["years"], [i * 100 for i in range(len(years))],'--', label="Dummy")

plt.xlabel("Tahun")
plt.ylabel("Dalam jutaan rupiah")
plt.title("Data Rekapitulasi Tahunan")
plt.legend()
plt.show()
```
