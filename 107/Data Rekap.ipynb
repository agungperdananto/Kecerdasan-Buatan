```python
# data_graph = {
#     "years": [],
#     "sales": []
# }
```

```python
# mapping data {tahun: penjualan}
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}
data_map
```

```python
# sorting years
years = list(data_map.keys())
years.sort()
years
```

```python
data_graph = {}
for year in years:
    sales_list = data_graph.get("sales",[])
    years_list = data_graph.get("years", [])
    years_list.append(year)
    sales_list.append(data_map[year])
    data_graph["sales"] = sales_list
    data_graph["years"] = years_list

data_graph
    
```

```python
plt.plot(data_graph["years"], data_graph["sales"], label="penjualan")
plt.plot(data_graph["years"], [i * 100 for i in range(len(years))], label="dummy")

plt.xlabel("Tahun")
plt.ylabel("Dalam nominal jutaan rupiah")
plt.title("Data Rekapitulasi")
plt.show()
```
