```python
# mapping data {tahun: penjualan}
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}
data_map
```

```python
# sorting years
years = list(data_map.keys())
years.sort()
years
```

```python
data_graph = {}
for year in years:
    data_graph_tahun = data_graph.get("tahun", [])
    data_graph_penjualan = data_graph.get("penjualan", [])
    data_graph_tahun.append(year)
    data_graph["tahun"] = data_graph_tahun
    data_graph_penjualan.append(data_map[year])
    data_graph["penjualan"] = data_graph_penjualan

data_graph
```

```python
plt.plot(data_graph["tahun"], data_graph["penjualan"], ".--", label="penjualan")
plt.plot(data_graph["tahun"], [i * 100 for i in range(7)], ".--", label="marketing")
plt.legend(loc="upper left")
plt.xlabel("Tahun")
plt.ylabel("Dalam Nominal Jutaan Rp.")
plt.title("Data Rekapitulasi")
plt.show()
```
