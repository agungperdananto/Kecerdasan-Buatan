```python
weight = {"hadir": 10, "tugas": 20, "uts":30, "uas":40}

def calculate(sks, hadir, tugas, uts, uas, bobot=weight):
    
    max_hadir = 0
    if sks == 2: 
        max_hadir = 14
    if sks == 3: 
        max_hadir = 21

    if not max_hadir:
        return None, "SKS tidak valid"
    hadir = hadir/max_hadir * bobot["hadir"]
    tugas = tugas/100 * bobot["tugas"]
    uts = uts/100 * bobot["uts"]
    uas = uas/100 * bobot["uas"]
    nilai = hadir + tugas + uts + uas
    return round(nilai,2), None
```

```python
tugas = 100
uts = 70
uas =70
hadir = 14
sks = 3

nilai, _err = calculate(sks, hadir,  tugas, uts, uas)
print(nilai, _err)

nilai, _err = calculate(tugas=tugas, hadir=hadir, sks=sks, uas=uas, uts=uts)
print(nilai, _err)
```
